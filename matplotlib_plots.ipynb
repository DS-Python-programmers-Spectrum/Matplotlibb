```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (assuming CSV format)
df = pd.read_csv('StudentsPerformance.csv')  # or 'student-por.csv'
```

```python
df
```

```python
# 1. Histogram of math scores
plt.hist(df['math score'], bins=20, edgecolor='black')
plt.title('Math Score Distribution')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# 6. Pie chart: gender distribution
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
# 7. Scatter plot: math vs reading scores
plt.scatter(df['math score'], df['reading score'])
plt.title('Math vs Reading Scores')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

```python
# 5. Bar chart: average reading score by parental education level
df.groupby('parental level of education')['reading score'].mean().plot(kind='bar')
plt.title('Average Reading Score by Parental Education')
plt.ylabel('Average Reading Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by race/ethnicity and compute average scores
grouped = df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()

# Line plot for average scores by race/ethnicity
plt.plot(grouped.index, grouped['math score'], marker='o', label='Math')
plt.plot(grouped.index, grouped['reading score'], marker='o', label='Reading')
plt.plot(grouped.index, grouped['writing score'], marker='o', label='Writing')

plt.title('Average Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Score')
plt.legend()
plt.grid(True)
plt.show()
```
